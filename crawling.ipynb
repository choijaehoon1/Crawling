```python
import pandas as pd 
import requests 

html = requests.get('https://finance.naver.com/sise/lastsearch2.nhn') 

table = pd.read_html(html.text)
```

```python
table[1]
```

```python
data = table[1].dropna().reset_index(drop = True)

data
```

```python
type(data['등락률'][0])
```

```python
for i in range(len(data)):
    if float(data['등락률'][i].replace('%', '')) < 0 :
        data['전일비'][i] = -data['전일비'][i]
        
data.columns = ['rank', 'issue', 'search_ratio', 'present_price', 'diff', 'diff_ratio', 'volume','open_price','high_price','low_price','per','roe']

data
```

```python
data.to_excel('stockpricesDB.xlsx')
```
